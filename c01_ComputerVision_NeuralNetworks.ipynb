## Mathematical model of an Artificial Neuron

Captura de pantalla 2023-12-04 114203.png

Inspirired by its biological couterpart (represented in the figure above), the artificial neuron takes several inputs (each a number), sums them together, and finally applies an activation function to obtain the output signal, which can be passed to the following neurons in the network (this can be seen as a directed graph) <br>
The summation of the inputs is usually done in a weighted way. Each **input** is scaled up or down, depending on a weight specific to this particular ***input**. These weights are the parameters that are adjusted during the training phase of the network in order for the neuron to react to the correct features. Often, another parameter is also trained and used for this summation process - the nueron's bias. Its value is simply added to the weighted sum as an offset

The **step function** is a key component of the original perceptron, but more advanced activation functions have been introduced since then with more advantageous properties, such as *non-linearity* (to model more complex behaviors) and *continuous differentiability* (important for the training process, which we wil explain later). The most common activation functions are as follows:
* The sigmoid function
* The hyperbolic tangent, tanh()
* The REctified Linear Unit (ReLU)
